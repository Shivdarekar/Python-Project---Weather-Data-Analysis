# Data Cleaning.

```python
df.isnull().sum()
```

```python
#Copy() --> Create same copy of the database.
data_clean = df.copy()
data_clean
```

```python
data_clean.info()
```

```python
# Change data type of the column "Temp C" --> int64, "Month" --> int64, "Weather" --> "category"

data_clean["Temp C"]= data_clean["Temp C"].astype("int64")
```

```python
data_clean.to_csv("data___.csv")
```

```python
data_clean.loc[10,"Temp C"] = 79
```

```python
# Change data type of the column "Temp C" --> int64, "Month" --> int64, "Weather" --> "category"

data_clean["Month"] = data_clean["Month"].astype("int64")
```

```python
data_clean.loc[23,"Month"] = 5
```

```python
# Change data type of the column "Temp C" --> int64, "Month" --> int64, "Weather" --> "category"

data_clean["Weather"] = data_clean["Weather"].astype("category")
```

```python
# Drop column from the data set "Unnamed: 0 ","Temp C"

data_clean= data_clean.drop(["Unnamed: 0","Temp C"],axis=1)
```

```python
data_clean
```

```python
# Duplicate check

data_clean[data_clean.duplicated()]
```

```python
# Drop duplicated --> to check the duplicated values.
data_clean[data_clean.duplicated()].shape
```

```python
# Drop Duplicates --> To remove the duplicates values.
data_clean = data_clean.drop_duplicates()
```

```python
data_clean
```

```python
# Reset Index --> To reset the column index number.
data_clean = data_clean.reset_index()
```

```python
data_clean
```

```python
# Rename --> To rename the column name.
data_clean = data_clean.rename({"Solar.R":"Solar"},axis=1)
data_clean
```

```python
# Fill Na --> fillna,ffill,bfill
# Measure of central tendency --> Mean,Median,Mode --> For numerical data use mean & Median, Mode --> for string or categorical data.

x = round(data_clean.Ozone.mean(),2)
```

```python
data_clean.isnull().sum()
```

```python
x
```

```python
y = data_clean.Solar.median()
```

```python
y
```

```python
z = data_clean.Weather.mode()
```

```python
z
```

```python
# Fillna --> Replace the Null values with specified values.
data_clean["Ozone"] = data_clean["Ozone"].fillna(x)
```

```python
data_clean["Solar"] = data_clean["Solar"].fillna(y)
```

```python
data_clean["Weather"] = data_clean["Weather"].fillna(z)
```

```python
data_clean["Weather"] = data_clean["Weather"].ffill()
```

```python
data_clean
```

```python
data_clean.isnull().sum()
```

# Data Visualization.
- Using Matplotlib & Seaborn visualization libraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Draw the Histogram

# figure fuction manage the width and height of the histogram.
plt.figure(figsize =(7,5))

# Hist fuction is to create the histogram . When you pass the particular columns.
plt.hist(x = "Ozone" , data = data_clean , width = 10 , color = "c")

# Title show the title of histogram and X label & Y label show the label on X&Y axis.
plt.title("Ozone Histogram Chart")
plt.xlabel("Ozone")
plt.ylabel("density")
plt.show()

# Show function is used to hide the additional data of the histogram.
plt.show()
```

```python
plt.figure(figsize=(5,5))

sns.displot(x = "Ozone", data = data_clean,kde = True,)

plt.title("Ozone Histogram Chart")
plt.xlabel("Ozone")
plt.ylabel("density")
plt.show()
```

```python
# figure --> To adjust the figure size.
plt.figure(figsize =(7,5))

# Displot --> It is a graph in seaborn library.
sns.displot(x = "Temp", data = data_clean, color = "b",kde = True)

plt.title("Temp Histogram Chart")
plt.xlabel("Temp")
plt.ylabel("density")
plt.show()
```

```python
groupA = data_clean["Ozone"].nunique()
groupB = data_clean["Solar"].nunique()
groupC = data_clean["Wind"].nunique()
groupD = data_clean["Temp"].nunique()


my_lables = ["Ozone","Solar","Wind","Temp"]
my_values = [groupA, groupB, groupC, groupD]

plt.pie(my_values, labels = my_lables, autopct= "%1.0f%%", explode = (0,0.05,0,0.1))
plt.title("Percent wise climate Data")

# legend --> It shows elements names in the graphs.
plt.legend(title = "Climate Name",loc = "center")
plt.show()
```
